```python
import pandas as pd
# load your uploaded file
df = pd.read_csv("/content/iris.csv")
df.head()
```

```python
df.head(10)
```

```python
print("Shape:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
categorical_cols = df.select_dtypes(include=['object','category']).columns
print("Numerical columns:", list(numerical_cols))
print("Categorical columns:", list(categorical_cols))
```

```python
stats = df[numerical_cols].agg(['mean','median','std'])
stats
```

```python
std_values = df[numerical_cols].std()
highest_variability_col = std_values.idxmax()

print("Standard deviation:\n", std_values)
print("\nColumn with highest variability:", highest_variability_col)
print(f"\nJustification: {highest_variability_col} has highest variability because it has the largest standard deviation.")
```
