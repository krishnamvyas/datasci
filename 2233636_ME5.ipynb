```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Load dataset
data = pd.DataFrame({
    "Hours":[1,2,3,4,5,6,7,8],
    "Score":[40,45,50,55,65,70,75,80]
})

# 2. Apply Linear Regression
X = data[["Hours"]]
y = data["Score"]

model = LinearRegression()
model.fit(X,y)

# 3. Plot regression line
plt.scatter(X,y)
plt.plot(X, model.predict(X))
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression")
plt.show()

# 4. Predict for 9 hours
pred = model.predict([[9]])
print("Predicted Score for 9 hours:", pred[0])

# 5. Evaluate model
y_pred = model.predict(X)
print("MSE:", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
```

```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Load dataset
data = pd.DataFrame({
    "Size":[1000,1200,1500,1800,2000],
    "Bedrooms":[2,3,3,4,4],
    "Age":[5,4,3,2,1],
    "Price":[200000,250000,300000,350000,400000]
})

# 2. Apply Multiple Linear Regression
X = data[["Size","Bedrooms","Age"]]
y = data["Price"]

model = LinearRegression()
model.fit(X,y)

# 3. Predict price for given values
pred = model.predict([[1600,3,2]])
print("Predicted House Price:", pred[0])

# 4. Evaluate model
y_pred = model.predict(X)
print("MSE:", mean_squared_error(y, y_pred))
print("R2 Score:", r2_score(y, y_pred))
```

```python
# Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Load dataset
data = pd.DataFrame({
    "Hours":[2,3,4,5,6,1,7,2],
    "Attendance":[60,65,70,75,80,50,85,55],
    "Result":["Fail","Fail","Pass","Pass","Pass","Fail","Pass","Fail"]
})

# 2. Convert Result to numeric
data["Result"] = data["Result"].map({"Fail":0, "Pass":1})

# Features and target
X = data[["Hours","Attendance"]]
y = data["Result"]

# 3. Train Decision Tree
model = DecisionTreeClassifier()
model.fit(X,y)

# 4. Predict for new student
pred = model.predict([[4,72]])
print("Prediction (0=Fail, 1=Pass):", pred[0])

# 5. Evaluate model
y_pred = model.predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```

```python
# Import libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Load dataset
data = pd.DataFrame({
    "Age":[22,25,30,35,40,28,45,23],
    "Income":[20000,25000,30000,35000,40000,27000,45000,22000],
    "Purchased":["No","No","Yes","Yes","Yes","No","Yes","No"]
})

# 2. Convert Purchased to numeric
data["Purchased"] = data["Purchased"].map({"No":0, "Yes":1})

# Features and target
X = data[["Age","Income"]]
y = data["Purchased"]

# 3. Train Random Forest
model = RandomForestClassifier()
model.fit(X,y)

# 4. Predict for new customer
pred = model.predict([[32,32000]])
print("Prediction (0=No, 1=Yes):", pred[0])

# 5. Evaluate model
y_pred = model.predict(X)

print("Accuracy:", accuracy_score(y, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```

```python
# Import library
from sklearn.metrics import confusion_matrix

# 1. Store actual and predicted values
actual = ["Spam","Spam","Spam","Not Spam","Not Spam","Spam","Not Spam","Spam"]
predicted = ["Spam","Not Spam","Spam","Spam","Not Spam","Spam","Not Spam","Not Spam"]

# Convert to numeric (Spam=1, Not Spam=0)
a = [1 if x=="Spam" else 0 for x in actual]
p = [1 if x=="Spam" else 0 for x in predicted]

# 2. Confusion Matrix
cm = confusion_matrix(a, p)
print("Confusion Matrix:\n", cm)

# 3. Extract values
TN, FP, FN, TP = cm.ravel()

print("TP:", TP)
print("TN:", TN)
print("FP:", FP)
print("FN:", FN)

# 4. Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

# 5. Display results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
